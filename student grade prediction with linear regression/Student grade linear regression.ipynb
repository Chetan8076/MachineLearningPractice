```python
df1 = df.copy()
```

```python
# df["GP"] = df["school"].apply(lambda x: 1 if x =="GP"  else 0)
# df["MS"] = df["school"].apply(lambda x: 1 if x =="MS"  else 0)
# df["MS"] = df["school"].apply(lambda x: 1 if x =="MS"  else 0)
```

```python
def fillValues(parentCol):
    for i in df1[parentCol].unique():
        df1[i] = df1[parentCol].apply(lambda x: 1 if x == i else 0)
```

```python
l = ["school","sex","address","famsize","Pstatus","Mjob","Fjob","reason","guardian"]
```

```python
YesNoValues = ["schoolsup","famsup","paid","activities","nursery","higher","internet","romantic"]
```

```python
for j in l:
    fillValues(j)

for j in YesNoValues:
    df1[j] = df1[j].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df1.drop(l,axis = 1,inplace=True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df1.corr())
```

```python
#spliting data on modified version of data that is df1
y = df1["G3"]
# p = ["G1","G2","G3"]

X =df1[[i for i in df1.columns if i!="G3" ] ]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X.head()
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
Predictions = lm.predict(X_test)
```

```python
plt.subplot(1,2,1)
plt.plot(Predictions)
plt.subplot(1,2,2)
plt.plot(y_test)
```

```python
plt.scatter(y_test,Predictions)
```

```python
from sklearn.metrics import mean_squared_error
```

```python

print(mean_squared_error(y_test,Predictions))
print(y_test.mean())

print(f'error :{mean_squared_error(y_test,Predictions)/y_test.mean()*100}%')
```
